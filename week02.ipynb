```python
colleges = pd.read_csv('data/college.csv', index_col='INSTNM')
# index_col 에서는 어떤 열을 인덱스로 사용할지 지정할 수 있습니다.
# INSTNM 은 학교 이름입니다.
college_ugds = colleges.filter(like='UGDS_')
college_ugds.head()
```

```python
# index_col 에서는 어떤 열을 인덱스로 사용할지 지정할 수 있습니다.
# pandas 는 숫자가 정 가운데, 0.5에 있으면 가까운 짝수 쪽으로 반올림을 시행합니다.
round(0.5 + 0.000001)
```

```python
college_ugds.filter(like='Northwest-Shoals', axis=0).T.add(.00001).round(2)
```
